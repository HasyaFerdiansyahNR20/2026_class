## Soal 1: Python Lists & Slicing
Diberikan sebuah list suhu dalam derajat Celsius: `temperatures = [25, 28, 30, 22, 35, 40, 18, 24]`

1. Gunakan **list comprehension** untuk membuat list baru bernama `temp_fahrenheit` yang mengonversi setiap suhu ke Fahrenheit. (Rumus: $F = (C \times 9/5) + 32$)
2. Ambil 3 data suhu terakhir dari list `temp_fahrenheit` menggunakan **slicing**.

````python
temperatures = [25, 28, 30, 22, 35, 40, 18, 24]

# Tulis jawaban Anda di bawah ini
temp_fahrenheit = [(c * 9/5) + 32 for c in temperatures]
print(temp_fahrenheit)

#3 data terakhir:
tiga_terakhir = temp_fahrenheit[-3:]
print(tiga_terakhir)
````

## Soal 2: Dictionary Operasi
Terdapat data stok barang di sebuah gudang dalam bentuk dictionary:
```python
inventory = {
    "laptop": {"stok": 10, "harga": 15000000},
    "mouse": {"stok": 50, "harga": 250000},
    "monitor": {"stok": 20, "harga": 3000000}
}
```
1. Tambahkan produk baru: `"keyboard"` dengan stok `30` dan harga `500000`.
2. Perbarui harga `"laptop"` menjadi `14500000` menggunakan metode `.update()`.
3. Hitung total nilai aset di gudang (stok dikali harga untuk semua barang).

````python
inventory = {
    "laptop": {"stok": 10, "harga": 15000000},
    "mouse": {"stok": 50, "harga": 250000},
    "monitor": {"stok": 20, "harga": 3000000}
}

# Tulis jawaban Anda di bawah ini
inventory["keyboard"] = {"stok": 30, "harga": 500000}
inventory["laptop"].update({"harga": 14500000})
total_aset = 0
for item in inventory:
    stok = inventory[item]["stok"]
    harga = inventory[item]["harga"]
    total_aset += stok * harga

print(f"Total nilai aset di gudang: Rp{total_aset:,}")
````

## Soal 3: NumPy Array & Aggregation
Diberikan matriks nilai mahasiswa (Baris = Mahasiswa, Kolom = Tugas 1, Tugas 2, Tugas 3):
```python
scores = np.array([
    [80, 85, 90],
    [70, 60, 75],
    [95, 90, 100],
    [40, 50, 45]
])
```
1. Hitung rata-rata nilai untuk **setiap mahasiswa** (axis=1).
2. Cari nilai tertinggi dari seluruh data menggunakan fungsi NumPy.
3. Gunakan `np.where` untuk membuat array baru yang berisi `"Lulus"` jika rata-rata mahasiswa >= 70 dan `"Gagal"` jika < 70.

````python
import numpy as np
scores = np.array([
    [80, 85, 90],
    [70, 60, 75],
    [95, 90, 100],
    [40, 50, 45]
])

# Tulis jawaban Anda di bawah ini

rata_rata = np.mean(scores, axis=1)
print("Rata-rata tiap mahasiswa:", rata_rata)


nilai_tertinggi = np.max(scores)
print("Nilai tertinggi di kelas:", nilai_tertinggi)


status = np.where(rata_rata >= 70, "Lulus", "Gagal")
print("Status kelulusan:", status)
````

## Soal 4: Pandas DataFrame Manipulation
Gunakan data di bawah ini untuk membuat DataFrame:
```python
df_data = {
    'Nama': ['Andi', 'Budi', 'Caca', 'Dedi', 'Euis'],
    'Departemen': ['IT', 'HR', 'IT', 'Sales', 'Sales'],
    'Gaji': [8000000, 7000000, 8500000, np.nan, 6000000],
    'Pengalaman_Tahun': [5, 3, 6, 2, 1]
}
```
1. Isi nilai Gaji yang kosong (NaN) dengan median dari seluruh gaji.
2. Buat kolom baru bernama `'Kategori_Senioritas'` dengan ketentuan: Jika pengalaman > 4 tahun maka `'Senior'`, jika tidak maka `'Junior'`. Gunakan fungsi `.apply()` atau `np.where`.
3. Tampilkan rata-rata Gaji per Departemen menggunakan `.groupby()`.

````python
import pandas as pd
import numpy as np

df_data = {
    'Nama': ['Andi', 'Budi', 'Caca', 'Dedi', 'Euis'],
    'Departemen': ['IT', 'HR', 'IT', 'Sales', 'Sales'],
    'Gaji': [8000000, 7000000, 8500000, np.nan, 6000000],
    'Pengalaman_Tahun': [5, 3, 6, 2, 1]
}

df = pd.DataFrame(df_data)

# Tulis jawaban Anda di bawah ini

df = pd.DataFrame(df_data)
median_gaji = df['Gaji'].median()
df['Gaji'] = df['Gaji'].fillna(median_gaji)
df['Kategori_Senioritas'] = np.where(df['Pengalaman_Tahun'] > 4, 'Senior', 'Junior')
rata_gaji_dept = df.groupby('Departemen')['Gaji'].mean()

print("DataFrame setelah Cleaning:")
print(df)
print("\nRata-rata Gaji per Departemen:")
print(rata_gaji_dept)
````
